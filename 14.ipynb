```python
import numpy as np
from PIL import Image

def average_images(image_list):
    # Mở ảnh đầu tiên và lấy kích thước của nó
    base_image = Image.open(image_list[0])
    base_size = base_image.size
    total_array = np.array(base_image, 'f')
    count = 1

    for image_path in image_list[1:]:
        try:
            image = Image.open(image_path)
            # Resize ảnh về kích thước của ảnh đầu tiên
            image = image.resize(base_size)
            image_array = np.array(image, 'f')
            total_array += image_array
            count += 1
        except Exception as e:
            print(f"Skip {image_path}, error: {e}")

    # Tính trung bình
    average_array = total_array / count

    # Chuyển đổi kết quả trung bình thành hình ảnh
    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image
```
