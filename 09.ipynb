```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'C:\\Users\\taithanhnguyen\\OneDrive - dntu.edu.vn\\Desktop\\ComputerVision\\Img\\04.jpg'
# Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# đây là code để chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
#biến ảnh thành số
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
#tạo viền cho các biên màu
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh
my_path = 'C:\\Users\\taithanhnguyen\\OneDrive - dntu.edu.vn\\Desktop\\ComputerVision\\Img\\04.jpg'

# đọc và hiển thị thành ảnh màu xám
im = Image.open(my_path).convert("L")

# Hiển thị lên trục
plt.imshow(im)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
